```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
data=pd.read_csv(r"D:\car_prices_updated.csv")
data
```

```python
dummy=pd.get_dummies(data['Car Model'],dtype=int)
dummy
```

```python
merged_data=pd.concat([data,dummy],axis='columns')
merged_data
```

```python
final=merged_data.drop(['Car Model','Toyota Corolla'],axis='columns')
final
```

```python
reg=linear_model.LinearRegression()
reg.fit(final[['Mileage','Age(yrs)','Audi A4','Audi A5','BMW X5','Mercedes C Class','Mercedez Benz C class']],data['Sell Price($)'])
```

```python
output=reg.predict([[2600,6,0,0,0,0,0]])
```

```python
output
```

```python
reg.score(final[['Mileage','Age(yrs)','Audi A4','Audi A5','BMW X5','Mercedes C Class','Mercedez Benz C class']],final['Sell Price($)'])
```
