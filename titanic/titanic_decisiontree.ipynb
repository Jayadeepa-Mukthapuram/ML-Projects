```python
import pandas as pd
```

```python
data=pd.read_csv(r"D:\titanic.csv")
data.head(30)
```

```python
inputs=data.drop(['PassengerId','Survived','Name','Ticket','Embarked','Cabin','SibSp','Parch'],axis='columns')

target=data['Survived']
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()

inputs['sex']=le_sex.fit_transform(inputs['Sex'])
inputs
```

```python
final_data=inputs.drop(['Sex'],axis='columns')
final_data
```

```python
mean_age=final_data['Age'].mean()
```

```python
final=final_data.fillna(mean_age)
final
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(final,target,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
